```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import os
```

```python
ANNOTATION_DIR = "/home/simon/Documents/Programming/Project3/VQGAN_Annotations/cleaned_annotations/"
```

```python
annotations = [ f.split(".")[0] for f in os.listdir(ANNOTATION_DIR)]
```

```python
annotations
```

```python
text_data = {}

for a in annotations:
    
    text_data[a] = {}
    
    with open(os.path.join(ANNOTATION_DIR, a + ".txt")) as f:
        
        for line in f:
            
            n, txt = line.split(": ")
            
            text_data[a][int(n)] = txt.strip()
```

```python
text_data.keys()
```

```python
text_data["17H13688_1A_IEC"][3]
```

```python
for a in text_data:
    
    for i in text_data[a]:
        
        string = np.array(text_data[a][i])
        
        np.save(os.path.join("./annotations", a + f"_{i}"), string)
        
```
